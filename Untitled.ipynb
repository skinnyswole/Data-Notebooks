```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/Users/pj/Documents/GitHub/Data-Notebooks/data/3Year_Projections.csv')
data.info()
clean = data.dropna()
clean
```

```python
for col in clean.columns[1:]: 
    clean[col] = clean[col].astype(str).str.replace('$', "").str.replace(',', "").str.replace('%', "") 
    clean[col] = pd.to_numeric(clean[col]) 
clean
```

```python
bar = sns.barplot(data=clean,
            x=clean.loc[0][1:],
            y=clean.columns)
bar
```
